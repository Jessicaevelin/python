# Parâmetro (Ativo, Passivo e DRE)

```
# Pacotes
import json  # Para ler o arquivo .json
import pandas as pd  # Para criar o DataFrame
import requests # Baixar o arquivo da internet
import numpy as np # Para exportar para excel
```

```
# Importar os dados diretamente do SES

# Parâmetros
ano = 2016 # ano dos dados contábeis
periodo = '2016-01-01'
# Importação do JSON (WebScraping)

# Ativo
url_ativo = f"http://dados.susep.gov.br/olinda-ide/servico/informacoescontabeis/versao/v1/odata/ContabeisAtivo(Ano=@Ano)?@Ano='2016'&$format=json"

# Documentação Request: https://requests.readthedocs.io/pt_BR/latest/user/quickstart.html
ativo = requests.get(url_ativo)
ativo = ativo.json()

# Transforma o dicionário data em lista obs.: Não consegui trabalhar bem com o 
# formato de dicionário, lista é mais fácil
dados_ativo = ativo['value']

# Passivo
url_passivo = f"http://dados.susep.gov.br/olinda-ide/servico/informacoescontabeis/versao/v1/odata/ContabeisPassivo(Ano=@Ano)?@Ano='2016'&$format=json"

# Documentação Request: https://requests.readthedocs.io/pt_BR/latest/user/quickstart.html
passivo = requests.get(url_passivo)
passivo = passivo.json()

# Transforma o dicionário data em lista obs.: Não consegui trabalhar bem com o 
# formato de dicionário, lista é mais fácil
dados_passivo = passivo['value']

# DRE
url_dre = f"http://dados.susep.gov.br/olinda-ide/servico/informacoescontabeis/versao/v1/odata/ContabeisDRE(Ano=@Ano)?@Ano='2016'&$format=json"

# Documentação Request: https://requests.readthedocs.io/pt_BR/latest/user/quickstart.html
dre = requests.get(url_dre)
dre = dre.json()

# Transforma o dicionário data em lista obs.: Não consegui trabalhar bem com o 
# formato de dicionário, lista é mais fácil
dados_dre = dre['value']
```

## Analise Financeira

### Indices de liquidez
Realizável Total = Ativo Circulante (+) Realizável a Longo Prazo;

Exigível Total = Passivo Circulante (+) Exigível a Longo Prazo.

Ativo Permanente = Investimentos + Imobilizado + Diferido

----
LIQUIDEZ GERAL = Realizável Total / Exigível Total

LIQUIDEZ CORRENTE = Ativo Circulante /
Passivo Circulante

LIQUIDEZ SECA = ( Ativo Circ. (–)Desp. Comerc. (–) Desp. Antec.) / Passivo Circulante

SOLVÊNCIA GERAL = Ativo Total / Exigível Total

ENDIVIDAMENTO = Exigível Total / Ativo Total

GARANTIA DE CAPITAIS DE TERCEIROS = Patrimônio Líquido / Exigível Total

IMOBILIZAÇÃO DO CAPITAL PRÓPRIO = Ativo Permanente / Patrimônio Líquido

INDEPENDÊNCIA FINANCEIRA = Patrimônio Líquido / Ativo Total

CAPITAL CIRCULANTE LÍQUIDO = Ativo Circulante (–) Passivo Circulante

LIQUIDEZ OPERACIONAL = Créditos Operacionais com Seguros / Débitos Operacionais com Seguros

COBERTURA VINCULADA = (Prov. Téc. Não Comprometidas (+) Compromas × 100 ) /  Ativo Total

```
# Ativo
ativo_circulante = df_ativo['CIRCULANTE']
realizavel_longoprazo = df_ativo['REALIZÁVEL A LONGO PRAZO']
realizavel_total = ativo_circulante + realizavel_longoprazo
despesa_antecipada = df_ativo.iloc[:,11]
ativo_total = df_ativo['TOTAL DO ATIVO']
ativo_permanente = df_ativo['INVESTIMENTOS'] + df_ativo['IMOBILIZADO'] + df_ativo['DIFERIDO']
credito_operacional_seguros = df_ativo.iloc[:,4]

# Passivo
passivo_circulante = df_passivo['CIRCULANTE']
exigivel_longoprazo = df_passivo['PASSIVO NÃO CIRCULANTE']
exigivel_total = df_passivo['CIRCULANTE'] + df_passivo['PASSIVO NÃO CIRCULANTE']
patrimonio_liquido = df_passivo['PATRIMÔNIO LÍQUIDO']
debito_operacional_seguros = df_passivo.iloc[:,3]
provisoes_tecnicas_comprometidas = df_passivo.iloc[:,7]
provisoes_tecnicas_nao_comprometidas = df_passivo.iloc[:,17]

# DRE
despesa_comercial = df_dre.iloc[:,8]
```

```
# indices financeiros
liquidez_geral = realizavel_total / exigivel_total

liquidez_corrente = ativo_circulante / passivo_circulante

liquidez_seca  = ((ativo_circulante - despesa_comercial - despesa_antecipada) / passivo_circulante) * 100

solvencia_geral = ativo_total / exigivel_total

endividamento = exigivel_total / ativo_total

garantia_capitais_terceiros = patrimonio_liquido / exigivel_total

imobilizacao_capital_proprio = ativo_permanente / exigivel_total

independencia_financeira = patrimonio_liquido / ativo_total

capital_circulante_liquido = ativo_circulante - passivo_circulante

liquidez_operacional = credito_operacional_seguros / debito_operacional_seguros

cobertura_vinculada = ((provisoes_tecnicas_comprometidas + provisoes_tecnicas_nao_comprometidas) / ativo_total ) * 100
```
