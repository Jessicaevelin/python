## Analise Financeira

### Indices de liquidez
Realizável Total = Ativo Circulante (+) Realizável a Longo Prazo;

Exigível Total = Passivo Circulante (+) Exigível a Longo Prazo.

Ativo Permanente = Investimentos + Imobilizado + Diferido

----
LIQUIDEZ GERAL = Realizável Total / Exigível Total

LIQUIDEZ CORRENTE = Ativo Circulante /
Passivo Circulante

LIQUIDEZ SECA = ( Ativo Circ. (–)Desp. Comerc. (–) Desp. Antec. × 100 ) / Passivo Circulante

SOLVÊNCIA GERAL = Ativo Total / Exigível Total

ENDIVIDAMENTO = Exigível Total / Ativo Total

GARANTIA DE CAPITAIS DE TERCEIROS = Patrimônio Líquido / Exigível Total

IMOBILIZAÇÃO DO CAPITAL PRÓPRIO = Ativo Permanente / Patrimônio Líquido

INDEPENDÊNCIA FINANCEIRA = Patrimônio Líquido / Ativo Total

CAPITAL CIRCULANTE LÍQUIDO = Ativo Circulante (–) Passivo Circulante

LIQUIDEZ OPERACIONAL = Créditos Operacionais com Seguros / Débitos Operacionais com Seguros

COBERTURA VINCULADA = (Prov. Téc. Não Comprometidas (+) Compromas × 100 ) /  Ativo Total
